```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
dtS = pd.read_csv('Bank_customer_retirement.csv')
```

```python
dtS
```

```python
dtS.drop('Customer ID', axis=1, inplace=True)
```

```python
dtS
```

```python
X = dtS.drop('Retire', axis=1)
y= dtS['Retire']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)
X_min = X_train.min()
X_range= (X_train-X_min).max()
X_scaled= (X_train-X_min)/X_range
```

```python
sns.scatterplot(x=X_scaled['Age'], y=X_scaled['401K Savings'], hue=y_train)
```

```python
X_train
```

```python
sns.scatterplot(x=X_train['Age'], y=X_train['401K Savings'], hue=y_train)
```

```python
from sklearn.svm import SVC

X_test_min = X_test.min()
X_test_range = (X_test-X_test_min).max()
X_test_scaled=(X_test-X_test_min)/X_test_range
param_grid = {'C':[1, 10, 100, 1000], 'gamma':[1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
grid.fit(X_scaled, y_train)
y_predict_po = grid.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict_po)
sns.heatmap(cm, annot=True, fmt= 'd')
```

```python
cls=SVC()
cls.fit(X_train, y_train)
y_predict = cls.predict(X_test)
cm=confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
cls_scaled=SVC()
cls_scaled.fit(X_train, y_train)
y_scaled_predict = cls_scaled.predict(X_test_scaled)
cm_scaled=confusion_matrix(y_test, y_scaled_predict)
sns.heatmap(cm_scaled, annot=True)
```

```python
print(classification_report(y_test, y_predict_po), classification_report(y_test, y_scaled_predict), classification_report(y_test, y_predict))
```
