```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], 'target'))
df
```

```python
df.head()
```

```python
sns.pairplot(df, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean smoothness'])
```

```python
sns.scatterplot(x='mean area', y='mean perimeter', hue='target', data=df)
```

```python
plt.figure(figsize = (25,15))
sns.heatmap(df.corr(), annot=True )
```

```python
X=df.drop('target', axis=1)
X
y=df['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state=5)
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
svm_cls=SVC(random_state=5)
svm_cls.fit(X_train, y_train)
```

```python
y_predict= svm_cls.predict(X_test)
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
X_min = X_train.min()
X_min
X_range= (X_train-X_min).max()
X_scaled= (X_train-X_min)/X_range
X_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python

sns.scatterplot(x = X_scaled['mean area'], y = X_scaled['mean smoothness'], hue = y_train)
```

```python
X_test_min = X_test.min()
X_test_range = (X_test-X_test_min).max()
X_test_scaled = (X_test-X_test_min)/X_test_range
sns.scatterplot(x=X_test['mean texture'], y=X_test['mean perimeter'], hue = y_test)
```

```python
sns.scatterplot(x=X_test_scaled['mean texture'], y=X_test_scaled['mean perimeter'], hue = y_test)
```

```python
svm_cls=SVC()
svm_cls.fit(X_scaled, y_train)
```

```python
y_predict2=svm_cls.predict(X_test_scaled)
cm2=confusion_matrix(y_test, y_predict2)
sns.heatmap(cm2, annot=True)
```

```python
print(classification_report(y_test, y_predict2))
```

```python
param_grid = {'C':[1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
grid.fit(X_scaled ,y_train)
```

```python
y_param_op_predict = grid.predict(X_test_scaled)
```

```python
cm=confusion_matrix( y_test, y_param_op_predict)
sns.heatmap(cm, annot=True, fmt= 'd')
```

```python
print(classification_report(y_test, y_param_op_predict))
```
